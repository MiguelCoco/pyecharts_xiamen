```python
from pyecharts import Map,Liquid
```

```python
#厦门市调查人口分布图
value = [274,848,495,1247,346,127]
attr = ['海沧区','湖里区','集美区','思明区','同安区','翔安区']
map = Map("厦门市调查人口分布图", width=1200, height=600)
map.add("", attr, value, maptype='厦门',is_label_show=True,label_text_color="#000",is_visualmap=True, 
        visual_text_color='#000',visual_range=[0,1300],)
map
```

```python
#厦门市男性占比水球图
sex_ratio = Liquid("男性占比")
sex_ratio.add("男性占比", [0.49],shape='pin')
sex_ratio
```

```python
#厦门市女性占比水球图
sex_ratio = Liquid("女性占比")
sex_ratio.add("女性占比", [0.51],shape='pin')
sex_ratio
```

```python
from pyecharts import Polar

radius = ['海沧区','湖里区','集美区','思明区','同安区','翔安区']
polar = Polar("厦门市各行政区拥有车辆分布", width=1200, height=600)
polar.add("私人小汽车", [169,375,299,518,222,72], radius_data=radius,
          type='barAngle', is_stack=True)
polar.add("摩托车", [85,4,301,7,289,114], radius_data=radius,
          type='barAngle', is_stack=True)
polar.add("电动车", [117,198,134,126,127,20], radius_data=radius,
          type='barAngle', is_stack=True)
polar.add("其他", [2,3,5,3,9,1], radius_data=radius,
          type='barAngle', is_stack=True)
polar
```

```python
from pyecharts import Scatter3D

import random
data = [
    [random.randint(0, 100),
    random.randint(0, 100),
    random.randint(0, 100)] for _ in range(80)
]
range_color = [
    '#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
    '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
scatter3D = Scatter3D("3D 散点图示例", width=1200, height=600)
scatter3D.add("", data, is_visualmap=True, visual_range_color=range_color)
scatter3D
```
